```python
#Importing Liblaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# **Load_Dataset**


```python
df=pd.read_csv('/content/Advertising.csv')
df=df.drop('Unnamed: 0',axis=1)
df.head(5)
```

# **Visualize Relationships Between All Variables**

```python
# Scatter plot of TV vs Sales with the regression line
sns.regplot(x='TV', y='Sales', data=df, line_kws={"color": "red"})
plt.title('TV Advertising Spend vs Sales with Regression Line')
plt.show()

# Scatter plot of TV vs Sales with the regression line
sns.regplot(x='Radio', y='Sales', data=df, line_kws={"color": "Blue"})
plt.title('Radio Advertising Spend vs Sales with Regression Line')
plt.show()

# Scatter plot of TV vs Sales with the regression line
sns.regplot(x='Newspaper', y='Sales', data=df, line_kws={"color": "Pink"})
plt.title('Newspaper Advertising Spend vs Sales with Regression Line')
plt.show()


```

# **Define features and target**

```python
X=df.drop('Sales',axis=1)
```

```python
y=df['Sales']
```

#**# Split Into train and test sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#**Train the model**

```python
#Train in a Linear Regression Model

from sklearn.linear_model import LinearRegression

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
```

# **Evaluating Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate MSE and R-squared
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

# **Using the Linear_Model For Future Predictions**

```python
# Example of new data-01
new_data = [[150, 30, 20]]  # TV = 150, Radio = 30, Newspaper = 20

# Predict using the trained model
predicted_sales = model.predict(new_data)
print(f'Predicted Sales: {predicted_sales[0]}')
```

```python
# Example of new data-2

# New data
new_data = [[300, 50, 50]]  # TV = 150, Radio = 30, Newspaper = 20

# Predict using the trained model
predicted_sales = model.predict(new_data)
print(f'Predicted Sales: {predicted_sales[0]}')
```
